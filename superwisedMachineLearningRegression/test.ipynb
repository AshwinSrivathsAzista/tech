```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
df = pd.read_csv("./data/cars.csv")
```

```python
df.corr()
```

```python
a = df.loc[:, 'Mileage' : 'Age'].copy()

a['temp'] = 1

a = a[['temp', 'Mileage', 'Engine_Capacity', 'Age']]

```

```python
y = df['Premium'].to_numpy()
# x = df.loc['Mileage', 'Engine_Capacity', 'Age']

# x = df.loc[:, 'Mileage' : 'Age'].to_numpy()
df['temp'] = 1

x = df[['temp', 'Mileage', 'Engine_Capacity', 'Age']].to_numpy()

# x
# x.to_numpy()
xT = x.transpose()
xTy = xT @ y
xTxInv = np.linalg.inv(xT @ x)
beta = xTxInv @ xTy

print('beta////////////////////////////////')
beta

print('y predicted ///////////////////////////////')
df['y'] = df['Premium (in dollars)']
df['yPredected'] = beta[0] + (df.Mileage * beta[1]) + (df.Engine_Capacity * beta[2]) + (df.Age * beta[3])
print(df.yPredected)

print('/////////////////////////////////////////')
yBar = df.y.mean()
print('y bar = ', yBar)

SSE = 0
SST = 0
for i in (df['y'] - df['yPredected']):
    SSE += i**2

for i in df.y:
    SST += (i - yBar) ** 2

print('SSE = ', SSE)
print('SST = ', SST)

rSq = 1 - (SSE/SST)
print('R square = ', rSq)
noOfObs = len(df.y)
print('number of observaitons = ', noOfObs)
mse = SSE/noOfObs
print('mse = ', mse)



# np.corrcoef(beta)



sb.scatterplot(y = y, x = x)
```
