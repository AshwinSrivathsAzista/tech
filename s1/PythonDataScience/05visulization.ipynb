Visulization using Matplotlib and Seaborn

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

df1 = pd.read_csv('lds_files_v1/lds41.csv')
df1
```

# line plot


```python
x1 = np.arange(1,6)
y1 = x1 * 2
y2 = x1 ** 2
print(x1)
print(y1)
print(y2)
```

```python
plt.plot(x1,y1, marker='*')
plt.plot(x1,y2, marker='D', color='red', label='expo')
plt.title('linear expo plot')

plt.xlabel('x_axis')
plt.ylabel('yaxis', rotation=45, fontsize=20, fontstyle='italic')
plt.legend()
plt.show()
```

```python
plt.plot(df1.wt,df1.mpg, marker='>')
df1.plot(x='wt', y='mpg', kind='line', marker='*')
```

```python
df1.iloc[:5].plot(x='wt', y='mpg', kind='line', marker='*')
```

```python
df1.sort_values(by='wt').plot(x='wt', y='mpg', kind='line', marker='*')


"""note
although the scatter points are the same and the scatter plots will be the same, the line graph is neater in the sorted case
"""
```

# scatter plot

```python
plt.scatter(x=x1, y=y1, label='lab')
```

```python
#  adding text to a plot
plt.scatter(x=x1, y=y1)

for xc,yc in zip(x1, y1):
    # print(xc, yc)
    plt.text(x=xc,y=yc,s=f"  ({xc},{yc})")
```

```python
v1=list(range(200,300))
v1.append(10)
v1.append(20)
v1.append(400)

plt.boxplot(v1)
```

```python
n1=np.array(v1)

q1 = np.percentile(n1,25)
q2 = np.percentile(n1,50)
q3 = np.percentile(n1,75)

iqr = q3 - q1

uwl = q3 + 1.5*iqr
lwl = q1 - 1.5*iqr

plt.text(1, q1, str(q1))
plt.text(1, q2, str(q2))
plt.text(1, uwl, 'UWL')
plt.text(1.2, lwl, 'lwl')

plt.axhline(uwl)
plt.axhline(lwl)

plt.boxplot(n1)
```
