# Ranking
          95 80 80 80  75
- dense:  1  2   2  2  3    -> no skip in rank
- first:  1  2   3  4  5    -> continuous numbers
- min:    1  2   2  2  5    -> dense but there is skip in rank
- max:    1  4   4  4  5    -> min but uses the highest value instead of the lowest
- avg:    1  3   3  3  5    -> average value used

```python
df1.Vol.rank()
df1.Vol.rank(ascending=False, method='average')
df1.Vol.rank(ascending=False, method='first')
df1.Vol.rank(ascending=False, method='dense')
df1.Vol.rank(ascending=False, method='min')
df1.Vol.rank(ascending=False, method='max')


```

# stack
- source level can be chosen but destination level is always -1 or closest to data
- topmost header is index 0 and the header closest to data is index -1

```python
col_level0_labels = [2022,2023]
col_level1_labels = ['q1','q2']
col_labels = pd.MultiIndex.from_product([col_level0_labels, col_level1_labels])

row_level0_labels = ['tata', 'birla', 'wipro', 'itc']
row_level1_labels = ['fmcg', 'tech']
row_labels = pd.MultiIndex.from_product([row_level0_labels, row_level1_labels])

df1 = pd.DataFrame(n84, index=row_labels, columns=col_labels)
df1

```

```python
df1.stack(future_stack=True)
# df1.unstack(1).unstack()
# df1.unstack(-1)

df1.stack().stack().unstack(0).unstack(-1)
```
