From a sample of 250 observations it is found that average income of a 27-year-old
Londoner is £45,000 with a sample standard deviation of £4000. Obtain the 95%
confidence interval to estimate the average income.

```python
n = 250000
std = 4000
conf_interval = 0.95
sample_mean = 45000

sample_mean_std = std/pow(n, 0.5)
print(stats.norm.interval(conf_interval, sample_mean, sample_mean_std))

# since we are using sample std deviation we should use t distribution
# but n is large and hence we can use normal distribution

stats.t.interval(conf_interval, n-1, sample_mean, sample_mean_std)
#  as n increases then it approaches normal distribution
```
