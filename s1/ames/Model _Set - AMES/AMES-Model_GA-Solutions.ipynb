#### Question 1  (Marks 5)

 The Edison Electric Institute has published figures on the number of kilowatt hoursused annually by various home appliances. It is claimed that a vacuum cleaner usesan average of 46 kilowatt hours per year. If a random sample of 12 homes included in a planned study indicates that vacuum cleaners use an average of 42 kilowatthours per year with a standard deviation of 11.9 kilowatt hours, does this suggest at the 0.05 level of significance that vacuum cleaners use, on average, less than 46kilowatt hours annually? Assume the population of kilowatt hours to be normal.
 

```python
u0 = 46
sampleMean = 42
sampleStdDev = 11.9

sampleSize = 12


h0 = "The population mean is 46"
h1 = "u0<46   The population mean is less than 46 as 42 is less than 46 and our observation is trying to disprove h0"

# Calculate the t-statistic and use cdf as I want the left tail

t = (sampleMean - u0) / (sampleStdDev / (sampleSize ** 0.5))
print(t)


p_value = stats.t.cdf(t, df=sampleSize - 1)
p_value
```

At the $0.05$ level of significance, there is not enough evidence to suggest that vacuum cleaners use less than $46$ kWh annually. The difference between $42$ and $46$ in this small sample could likely be due to random chance.

```python
# Mean = 46, n=12, Sigma = 11.9
# HO : Mu =46
# H1 : Mu < 46
# Data Mu = 46, Xbar = 42, n=12, Sigma = 11.9
tcalc= (42-46)/(11.9/np.sqrt(12))
tcalc
```

```python
def PCA(X , num_components):
     
    #Step-1
    X_meaned = X - np.mean(X , axis = 0)
     
    #Step-2
    cov_mat = np.cov(X_meaned , rowvar = False)
     
    #Step-3
    eigen_values , eigen_vectors = np.linalg.eigh(cov_mat)
     
    #Step-4
    sorted_index = np.argsort(eigen_values)[::-1]
    sorted_eigenvalue = eigen_values[sorted_index]
    sorted_eigenvectors = eigen_vectors[:,sorted_index]
     
    #Step-5
    eigenvector_subset = sorted_eigenvectors[:,0:num_components]
     
    #Step-6
    X_reduced = np.dot(eigenvector_subset.transpose() , X_meaned.transpose() ).transpose()
     
    return X_reduced
```
