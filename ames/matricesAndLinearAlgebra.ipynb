```python
a = np.array([[1, 2, 3], [8, 7, 3]])
b = np.array([[1, 2, 9], [4, 7, 1]])


print('a.shape => ', a.shape)
print('a + b => ', a + b)
print('a - b => ', a - b)
print('a * b (not matrix multiplication and pointwise multiplication) => ', a * b) # not matrix multiplication and pointwise multiplication
print('a @ b.T (matrix multiplication) ', a @ b.T) # matrix multiplication
print(np.matmul(a, b.T)) # matrix multiplication
```

```python
"""
"""
print(np.linalg.solve([[2, 3, 4], [4, 6, 0], [0, 1, 9]], [10, 12, 18]))

a = np.array([[2, 3, 4], [4, 6, 0], [0, 1, 9]])
c = np.array([10, 12, 18])

print(np.linalg.solve(a, c))

print(np.linalg.inv(a) @ c)

aInv = np.linalg.inv(a)
np.matmul(aInv, c)
```

# determinant
- gives volume of space created by the vectors
- if determinant is 0 then the vectors are dependent
- independent vectors are vectors that cant be derived by a linear combination of a subset of the vectors

```python
a = [1,2]
b = [2,3]
c = np.array([a, b])

np.linalg.det(c)
```

```python
np.linalg.det([[1,2], [2,4]]) # no space of volume and the vectors are dependent on each other => [1,2] * 2 = [2,4]
```

```python
np.linalg.det(np.array([[1, 2, 3], [4, 0, 2], [5, 2, 5]]))
```

# Limits

```python
# import sympy as sym
x = sym.Symbol('x')
y = sym.Symbol('y')

print(x)
print(x**2 + 2*x + 1)
print(sym.expand((x + 1)**2))
print(sym.factor(x**2 + 2*x + 1))
print(sym.solve(x**2 + 2*x + 1, x))
print(sym.limit((x**2 - 1)/(x - 1), x, 1))
print(sym.simplify((x**2 - 1)/(x - 1)))
```

```python
x = np.arange(-4*np.pi, 4*np.pi, 0.1) # starting from -4π to 4π with a step of 0.1
y = np.sin(x)/x


plt.plot(x, y)
plt.show()
```

```python
x = sym.Symbol('x')
y = sym.Symbol('y')
sym.limit((sym.sin(x) / x), x, 0)
```

```python
sym.limit(sym.log(x) / x, x, 0)
```

```python
sym.limit((sym.E**x - 1) / x, x, 0)
```

```python
# derivatives
sym.diff(sym.sin(x**2), x)
```

```python
 # second derivative -> substutitute x=4 and evaluate
sym.diff(sym.sin(x**2), x, 2).subs(x, 4).evalf() 
```

```python
# jacobian input multiple output multiple dimension
f = sym.Matrix([x**2, x+y])
f.jacobian([x,y])
```

```python
f = sym.Matrix([sym.sin(x+y), sym.cos(x+y)])
f.jacobian([x,y])
```

```python
# hassian
# input multiple output single dimension

x,y = sym.symbols('x y')
eq = -7*(x**5) - 5*x*y - 4*(y**6)
# v = list(ordered(f.free_symbols))

sym.hessian(eq, [x,y])
```

```python
f = x**3 - 3*x**2 + 2

fDash = sym.diff(f, x)
fDoubleDash = sym.diff(f, x, 2)

critical_points = sym.solveset(fDash, x, domain=sym.Reals)
critical_points
```
